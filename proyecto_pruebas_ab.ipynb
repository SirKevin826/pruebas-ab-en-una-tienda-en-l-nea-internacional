La mayor pérdida de usuarios ocurre en la etapa Page → Cart:

* Grupo A: Pierde 2,693 usuarios (52.3%)
* Grupo B: Pierde 1,836 usuarios (48.6%)

**La Anomalía Principal**
Se descubrió que la tasa de conversión de Cart → Purchase es del 108.2% en el Grupo A y 98.4% en el Grupo B. Esto significa que más usuarios compraron que los que agregaron productos al carrito, lo cual parece contradictorio.

¿Por qué ocurre esto?
La explicación principal es que los usuarios pueden comprar sin necesariamente agregar productos al carrito primero. Esto puede suceder por varias razones:

1. Compra Directa
* Algunos usuarios pueden tener la opción de "Comprar Ahora" que evita el carrito
* Productos con compra de un solo clic
* Usuarios que usan funciones de compra rápida

2. Comportamiento de Usuario Diverso
* Usuarios que navegan, salen, y regresan más tarde para comprar directamente
* Usuarios que usan diferentes dispositivos o sesiones
* Compras a través de enlaces directos o promociones

3. Limitaciones en el Tracking
* Posibles gaps en el seguimiento de eventos
* Usuarios que limpian cookies entre sesiones
* Diferentes flujos de compra no capturados completamente

## Prueba Z

Ahora que se han implementado la estrategia de rescate con datos balanceados, se hizo la prueba z a conversión total

```python
# Extraer datos del embudo balanceado
grupo_A_bal = embudo_balanceado.loc['A']
grupo_B_bal = embudo_balanceado.loc['B']

# Datos para la prueba Z (Page → Purchase)
purchase_A_bal = grupo_A_bal['purchase']
page_A_bal = grupo_A_bal['product_page']
purchase_B_bal = grupo_B_bal['purchase']
page_B_bal = grupo_B_bal['product_page']

print(f"=== DATOS PARA PRUEBA Z BALANCEADA ===")
print(f"Grupo A: {purchase_A_bal}/{page_A_bal} = {purchase_A_bal/page_A_bal:.3f}")
print(f"Grupo B: {purchase_B_bal}/{page_B_bal} = {purchase_B_bal/page_B_bal:.3f}")
```

```python
import numpy as np
from scipy import stats

def prueba_z_proporciones(x1, n1, x2, n2):
    """
    x1, x2: número de éxitos en cada grupo
    n1, n2: tamaño de muestra de cada grupo
    """
    # Proporciones
    p1 = x1 / n1
    p2 = x2 / n2
    
    # Proporción combinada
    p_combined = (x1 + x2) / (n1 + n2)
    
    # Error estándar
    se = np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
    
    # Estadístico z
    z = (p1 - p2) / se
    
    # Valor p (prueba de dos colas)
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    
    return z, p_value, p1, p2
```

```python
# Aplicar la prueba Z balanceada
z_stat_bal, p_value_bal, prop_A_bal, prop_B_bal = prueba_z_proporciones(
    purchase_A_bal, page_A_bal,
    purchase_B_bal, page_B_bal
)

print("\n=== RESULTADOS DE LA PRUEBA Z BALANCEADA ===")
print(f"Estadístico z: {z_stat_bal:.4f}")
print(f"Valor p: {p_value_bal:.6f}")
print(f"Proporción Grupo A: {prop_A_bal:.4f} ({prop_A_bal*100:.1f}%)")
print(f"Proporción Grupo B: {prop_B_bal:.4f} ({prop_B_bal*100:.1f}%)")
print(f"Diferencia: {(prop_A_bal - prop_B_bal)*100:.1f} puntos porcentuales")
```

```python
# Interpretación del resultado
alpha = 0.05
print(f"\n=== INTERPRETACIÓN (α = {alpha}) ===")
if p_value_bal < alpha:
    print("RESULTADO: Diferencia ESTADÍSTICAMENTE SIGNIFICATIVA")
    print("Rechazamos la hipótesis nula")
    print("El nuevo sistema de recomendaciones SÍ tiene un efecto significativo")
else:
    print("RESULTADO: Diferencia NO estadísticamente significativa")
    print("No podemos rechazar la hipótesis nula")
    print("No hay evidencia suficiente de que el nuevo sistema sea mejor")

print(f"\nConfianza: {(1-alpha)*100}%")
```

```python
# Calcular el tamaño del efecto
diferencia_absoluta = abs(prop_A_bal - prop_B_bal)
diferencia_relativa = (diferencia_absoluta / prop_A_bal) * 100

print(f"\n=== ANÁLISIS DE SIGNIFICANCIA PRÁCTICA ===")
print(f"Diferencia absoluta: {diferencia_absoluta:.4f} ({diferencia_absoluta*100:.1f} puntos porcentuales)")
print(f"Diferencia relativa: {diferencia_relativa:.1f}%")

# Tamaño del efecto (Cohen's h para proporciones)
import numpy as np

def cohens_h(p1, p2):
    """Calcula Cohen's h para diferencia entre proporciones"""
    return 2 * (np.arcsin(np.sqrt(p1)) - np.arcsin(np.sqrt(p2)))

h = abs(cohens_h(prop_A_bal, prop_B_bal))
print(f"Cohen's h (tamaño del efecto): {h:.4f}")

# Interpretación del tamaño del efecto
print(f"\n=== INTERPRETACIÓN DEL TAMAÑO DEL EFECTO ===")
if h < 0.2:
    efecto = "TRIVIAL"
elif h < 0.5:
    efecto = "PEQUEÑO"
elif h < 0.8:
    efecto = "MEDIANO"
else:
    efecto = "GRANDE"

print(f"Tamaño del efecto: {efecto}")
```

```python
# Calcular el poder estadístico necesario para detectar esta diferencia
from scipy import stats

def calcular_tamano_muestra_necesario(p1, p2, alpha=0.05, power=0.8):
    """Calcula el tamaño de muestra necesario para detectar una diferencia"""
    effect_size = cohens_h(p1, p2)
    z_alpha = stats.norm.ppf(1 - alpha/2)
    z_beta = stats.norm.ppf(power)
    
    n = ((z_alpha + z_beta) / effect_size) ** 2
    return int(np.ceil(n))

n_necesario = calcular_tamano_muestra_necesario(prop_A_bal, prop_B_bal)
n_actual = min(page_A_bal, page_B_bal)

print(f"\n=== ANÁLISIS DE PODER ESTADÍSTICO ===")
print(f"Tamaño de muestra actual: {n_actual:,} por grupo")
print(f"Tamaño necesario para detectar esta diferencia: {n_necesario:,} por grupo")
print(f"Ratio actual/necesario: {n_actual/n_necesario:.2f}")

if n_actual < n_necesario:
    print("La muestra actual es INSUFICIENTE para detectar esta diferencia")
else:
    print("La muestra actual es suficiente")
```
